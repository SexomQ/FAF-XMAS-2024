```python
# create a tuple of all the coordinates from the dataframe like ((x1, y1), (x2, y2), ...)
coordinates = list(zip(df_santa['X Coordinate'], df_santa['Y Coordinate']))
print(coordinates[:5])
```

```python
import numpy as np

def haversine_distance_vectorized(coords1, coords2):
    """
    Calculate the straight-line distances (in meters) between multiple geographic points using the Haversine formula.
    
    Args:
        coords1: Array of shape (n, 2) where each row is (lat, lon) in decimal degrees.
        coords2: Array of shape (n, 2) where each row is (lat, lon) in decimal degrees.

    Returns:
        Distances in meters as a 1D NumPy array.
    """
    R = 6378137  # Radius of Earth in meters

    # Convert latitude and longitude to radians
    lat1, lon1 = np.radians(coords1[:, 0]), np.radians(coords1[:, 1])
    lat2, lon2 = np.radians(coords2[:, 0]), np.radians(coords2[:, 1])

    # Compute differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    # Distance in meters
    distances = R * c
    return distances


distances = np.zeros((len(coordinates), len(coordinates)))
# calculate the distance between all the coordinates
for i in range(len(coordinates)):
    distances[i] = haversine_distance_vectorized(np.array([coordinates[i]]), np.array(coordinates))

print(distances)
```

```python
import numpy as np
from geopy.geocoders import Nominatim
import requests
import json
import folium
import dotenv
import os
```

```python
class RouteVisualizer:
    def __init__(self, df=None, ordered_list=None):
        self.df = df
        self.ordered_list = ordered_list

    def plot_route(self, ordered_list):
        """
        Plot the route on a map using Folium.
        
        Args:
            ordered_list: List of indices representing the order of the route.
        """
        # Get the coordinates of the locations
        coords = self.df[['X Coordinate', 'Y Coordinate']].values
        coords = coords[ordered_list]

        # Get the center of the map
        center = coords.mean(axis=0)

        # Create the map
        m = folium.Map(location=center, zoom_start=12)

        # Add markers for each location
        for i, coord in enumerate(coords):
            folium.Marker(coord, popup=str(i)).add_to(m)

        # Add lines between the locations
        folium.PolyLine(coords, color="red", weight=2.5, opacity=1).add_to(m)

        return m
```

```python
route_visualizer = RouteVisualizer(df_santa, ordered_route)
m = route_visualizer.plot_route(ordered_route)
m.save('route.html')
```
